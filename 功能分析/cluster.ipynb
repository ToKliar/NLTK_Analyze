NLTK内部封装实现了三种常见的聚类方法： 
+ KMeans（K均值）
+ 混合高斯聚类（EM）
+ GAAC（凝聚层次聚类法）

这里需要注意 聚类器传入数据的要求是numpy的array类型

这里使用三种聚类器对NLTK内置语料库中的电影评论进行聚类

```python
import nltk
from numpy import array
from nltk.cluster import *


datas = [array(v) for v in [(1, 0), (0, 1), (1, 1), (5, 5), (5, 4), (4, 5)]];


kmeans_clus = KMeansClusterer(2, euclidean_distance) #分为两类，距离函数用欧拉距离
kmeans_clus.cluster(datas, True, trace=False)
print([str(kmeans_clus.classify(data)) for data in datas])

```

```python
em_clus = EMClusterer(initial_means=[[4, 2], [2, 1]])
em_clus.cluster(vectors = datas)
print([str(em_clus.classify(data)) for data in datas[:10]])
```

```python
gaac_clus = GAAClusterer(num_clusters=2, normalise=True)
gaac_clus.cluster(vectors=datas)
gaac_clus.dendrogram().show()
print([str(gaac_clus.classify(data)) for data in datas])
```
